```python
import string
import math
import pandas as pd

n = 12
freq = {}
tf = {}
df = {}
idf = {}
tf_idf = {}
termos_presentes = set()

for i in range(1, n + 1):
    with open(f'/content/drive/MyDrive/Colab Notebooks/Mineração de texto/{i}.txt', 'r') as arquivo:
        texto = arquivo.read()
        texto = texto.split(" ")
        termos_presentes.clear()
        for termo in texto:
            for pontuacao in string.punctuation:
                termo = termo.replace(pontuacao,'')
            termo = termo.replace(' \n','')
            termo = termo.replace('\n','')
            termo = termo.replace('“','')
            termo = termo.replace('”','')
            termo = termo.lower()
            termos_presentes.add(termo)
            if (termo, i) not in freq:
                freq[termo, i] = 1
            else:
                freq[termo, i] += 1

        for termo, doc in freq:
            if freq[termo, doc] > 0:
                tf[termo, doc] = round(1 + math.log10(freq[termo, doc]), 3)
            else:
                tf[termo, doc] = 0

        for palavra in termos_presentes:
            if palavra not in df:
                df[palavra] = 1
                idf[palavra] = round(math.log10(n), 3)
            else:
                df[palavra] += 1
                idf[palavra] = round(math.log10(n/df[palavra]), 3)

        for termo, doc in tf:
            if tf[termo, doc] > 0:
                tf_idf[termo, doc] = round(tf[termo, doc] * idf[termo], 3)
            else:
                tf_idf[termo, doc] = 0

freq = dict(sorted(freq.items()))
tf = dict(sorted(tf.items()))
df = dict(sorted(df.items()))
idf = dict(sorted(idf.items()))
```

```python
dataframe_df = pd.DataFrame({'Palavra': list(df.keys()), 'DF':list(df.values())})
dataframe_idf = pd.DataFrame({'Palavra': list(idf.keys()), 'IDF':list(idf.values())})
dataframe_df_idf = dataframe_df.merge(dataframe_idf, on='Palavra')
dataframe_df_idf
```

```python
dataframe_freq = pd.DataFrame(list(freq.values()), index=freq.keys())
dataframe_freq = dataframe_freq.unstack(fill_value=0)
dataframe_freq
```

```python
dataframe_tf = pd.DataFrame(list(tf.values()), index=freq.keys())
dataframe_tf = dataframe_tf.unstack(fill_value=0)
dataframe_tf
```

```python
dataframe_tf_idf = pd.DataFrame(list(tf_idf.values()), index=tf_idf.keys())
dataframe_tf_idf = dataframe_tf_idf.unstack(fill_value=0)
dataframe_tf_idf
```

```python
similaridade = {}

for i in range(0, n):
    for j in range(i + 1, n):
        cos = 0
        produto = 0
        A = 0
        B = 0

        for k in range(len(dataframe_tf_idf)):
            produto += (dataframe_tf_idf.iloc[k, i]) * (dataframe_tf_idf.iloc[k, j])
            A += (dataframe_tf_idf.iloc[k, i] ** 2)
            B += (dataframe_tf_idf.iloc[k, j] ** 2)

        if A > 0 and B > 0:
            cos = produto / (math.sqrt(A) * math.sqrt(B))

        documento_i = i + 1
        documento_j = j + 1
        similaridade[f'{documento_i}.txt - {documento_j}.txt -'] = cos

print('Os 4 documentos mais parecidos usando similaridade por cosseno:')
for i in sorted(similaridade, key=similaridade.get, reverse=True)[:4]:
    print(i, round(similaridade[i], 3))
```

```python
dataframe_df_idf.to_csv("Palavra x DF x IDF.csv", decimal=',')
dataframe_tf.to_csv("TF x Documento.csv", decimal=',')
dataframe_tf_idf.to_csv("TF x IDF.csv", decimal=',')
```
